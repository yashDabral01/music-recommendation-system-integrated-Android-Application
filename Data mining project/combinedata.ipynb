```python
import pandas as pd
```

```python
tracks_df = pd.read_csv('Music info.csv')  # Excel file with 'id' and 'track_name'
lyrics_df = pd.read_csv('spotify_millsongdata.csv')  # Excel file with 'track_name' and 'lyrics'

# Merge the two dataframes on the 'track_name' column
merged_df = pd.merge(tracks_df, lyrics_df, on='name', how='inner')

# Save the merged dataframe to a new Excel file
merged_df.to_csv('merged_tracks_lyrics.csv.', index=False)

print("Merged file has been created successfully!")
```
